```python
import tensorflow as tf
import matplotlib.pyplot as mp
import numpy as np

sn = 200

x = np.linspace(0, 2*np.pi, sn, dtype=np.float32)
mp.plot (x, np.sin(x**2))
mp.title ("A simple chirp - " + str(sn))

sess = tf.Session()
init = tf.global_variables_initializer()

sess.run(init)

mp.show()
```
